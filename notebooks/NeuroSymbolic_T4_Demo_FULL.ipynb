## 2. Import and Initialize System

```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
import time
from neurosymbolic import NeurosymbolicSystem

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

model = NeurosymbolicSystem().to(device)
model.eval()
print("✓ Model initialized")
```

## 3. Neural Perception Demo

```python
image = torch.randn(1, 3, 224, 224).to(device)
with torch.no_grad():
    output = model.perceive(image, threshold=0.5)

symbolic_scene = output["symbolic"][0]
print(f"Detected {len(symbolic_scene)} concepts:")
for concept, conf, attrs in symbolic_scene[:10]:
    print(f"  - {concept}: {conf:.3f} (attributes: {attrs})")
```

## 4. Symbolic Reasoning Demo

```python
reasoning_output = model.reason(symbolic_scene)
print(f"Derived {reasoning_output['num_derived']} new facts")
print("Top derived facts:")
for pred, args, conf, source in reasoning_output['derived_facts'][:10]:
    print(f"  - {pred}({args}) [conf: {conf:.3f}, source: {source}]")
```

## 5. Query-Based Inference

```python
query = ("dangerous", ("obj1",))
proofs = model.query(image, query)
if proofs:
    print(f"Query result: True (confidence: {proofs[0]['confidence']:.3f})")
    print("Proof steps:")
    for step in proofs[0]['proof']:
        print(f"  - {step}")
else:
    print("Query result: False")
```

## 6. Explanation Generation

```python
if reasoning_output['derived_facts']:
    fact = reasoning_output['derived_facts'][0]
    explanations = model.explain_prediction(image, (fact[0], fact[1]))
    for exp in explanations:
        print(exp)
```

## 7. Custom Rules

```python
model.reasoner.add_rule(
    head=("alert", ("?x",)),
    body=[("dangerous", ("?x",)), ("near", ("?x", "person"))],
    confidence=0.99
)
print("Custom rule added")
new_reasoning = model.reason(symbolic_scene)
print(f"New facts derived: {new_reasoning['num_derived']}")
```

## 8. Performance Benchmarking

```python
times = []
for _ in tqdm(range(50)):
    start = time.time()
    _ = model.forward(image)
    if torch.cuda.is_available(): torch.cuda.synchronize()
    times.append(time.time() - start)
print(f"Average Latency: {np.mean(times)*1000:.2f} ms")
print(f"Throughput: {1/np.mean(times):.2f} FPS")
```
